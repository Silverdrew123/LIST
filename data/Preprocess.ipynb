```python
data_dir = '/home/yaqing/Projects/Few-shot/PromptST/data/clue'
```

```python
task = 'RTE'
K=100
```

```python
for seed in [1,2,3,4,5]:
    path = os.path.join(data_dir, task)
    path = os.path.join(path, str(K)+'-'+str(seed))
    lines = DataProcessor._read_tsv(os.path.join(path, 'train.tsv'))
    header = lines[0]
    data = lines[1:]
    random.shuffle(data)

    f_train = open(os.path.join(path, 'train.tsv'), 'w')
    if task == 'MNLI':
        f_dev = open(os.path.join(path, 'dev_matched.tsv'), 'w')
    else:
        f_dev = open(os.path.join(path, 'dev.tsv'), 'w')
    f_train.write('\t'.join(header)+'\n')
    f_dev.write('\t'.join(header)+'\n')
    for l in data[:80]:
        f_train.write('\t'.join(l)+'\n')
    f_train.close()
    for l in data[80:]:
        f_dev.write('\t'.join(l)+'\n')
    f_dev.close()
        
```
